```python
import os

from ultralytics import YOLO

model = YOLO("yolov8x.pt")
projectPath = '/content/gdrive/MyDrive/Google Colab/YOLOv8x Epoch 100 Batch 32'

results = model.train(#model=model,
                      data=os.path.join(ROOT_DIR,"data.yaml"),
                      batch=32,
                      imgsz=640,
                      epochs=100,
                      project = projectPath,
                      name = 'YOLOv8x Epoch 100 Batch 32 with Training Plots',
                      optimizer = 'AdamW',
                      val = True,
                      plots = True
                      )
```
